### Subplots
---
**What is subplot in Matplotlib?**  
- A subplot in Matplotlib allows you to create multiple plots in a single figure. It helps in organizing multiple graphs neatly in a grid layout.

**1️⃣ Using plt.subplot() (MATLAB-style)**  
- plt.subplot(nrows, ncols, index)
- nrows: Number of rows in the grid
- ncols: Number of columns in the grid
- index: Position of the current subplot (starting from 1, left to right, top to bottom)

````python
# creating the layout
plt.figure(figsize = (12,6))
# naming the plot
plt.title("Learning subplot")
# creating subplot
plt.subplot(1,2,1)
# here index is 1 hence 1st subplot active
# ===================================================plotting in subplot 1
plt.plot(x, x ** 2)
# this line mean plotting points (0,-10) , (0,0) , (0,100)
plt.plot([0, 0, 0], [-10, 0, 100], marker='o', markersize=8, linestyle='-', color='green')

# plt.legend() is used to display a legend in a plot, which helps identify different plotted data series. 
# The legend provides labels for different lines, bars, or other plotted elements, making the graph more readable.
# In simple words legend helps to create keys for the plotting [like what signifies what in the plot]
plt.legend(['X^2', 'Vertical Line'])
plt.xlabel('X')
plt.ylabel('X Squared')

# making 2nd subplot active
plt.subplot(1, 2, 2)
# ===================================================plotting in subplot 2
plt.plot(x, -1 * (x ** 2))
plt.plot([-10, 0, 10], [-50, -50, -50], marker='o', markersize=8, linestyle='-', color='red')
plt.legend(['-X^2', 'Horizontal Line'])
plt.xlabel('X')
plt.ylabel('X Squared')
# these markers in the .plot is used to visualize the points in the line 
````

### OOP Interface

- Uses Figure and Axes objects explicitly.
- More flexible and better for creating multiple subplots or custom layouts.
- Recommended for complex plots and reusable code.

**Understanding fig and ax in Matplotlib (Figure & Axes Objects)**  
In the Object-Oriented (OO) Approach of Matplotlib, the two key components are:

**1️⃣ Figure (fig)**
- The entire canvas or container that holds all plots (subplots, texts, titles, etc.).
- Think of it as a blank sheet of paper where you can draw multiple graphs.

**2️⃣ Axes (ax)**
- Represents an individual plot (subplot) within the Figure.
- Each Axes object contains:
  - The actual plot
  - Labels (ax.set_xlabel(), ax.set_ylabel())
  - Titles (ax.set_title())
  - Legends (ax.legend())
  - Grid, ticks, etc.



````python
fig, axes = plt.subplots(figsize = (12,6))
# ✅ This returns a tuple: (fig, ax), where:
# fig (Figure object) → The entire figure (canvas).
# ax (Axes object) → A single subplot where we can plot data.
axes.plot(x,x**2, color = "red", linewidth = 3, marker = "o", markersize = 8, label = 'x^2')
axes.plot(x,-1*x**2, 'b--', label = '-x^2')

# creating legend (the marker for the plot)
axes.legend()

axes.set_xlabel('X')
axes.set_ylabel('X Squared')
axes.set_title("My Nice Plot")
````

````python
# hence fig contain the entire content of the plot along with all the subplots on the main plot 
fig
````

````python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, linestyle='solid')
axes.plot(x, x + 1, linestyle='dashed')
axes.plot(x, x + 2, linestyle='dashdot')
axes.plot(x, x + 3, linestyle='dotted');

axes.set_title("My Nice Plot")
````

### Other types of plots

#### Figures and subfigures

When we call the subplots() function we get a tuple containing a Figure and a axes element.

#### 🖼️ 1. Figure (Top-Level Container)

- A Figure is the entire canvas for your plot(s).
- It can hold one or more subplots (Axes).
- Think of it as a sheet of paper where you draw your graphs.

#### 🧩 2. SubFigure (New in Matplotlib 3.4+)  

- A SubFigure is a division of a Figure into multiple independent sections.
- Each SubFigure can contain its own Axes, titles, labels, and layout.
- Useful for more advanced layouts — for example, combining multiple distinct plot grids into one figure.
- 🧠 Each subfig acts like a mini-figure inside the main figure!

**QUESTION : Isnt there something like the subfigure called subplot ? Whats the difference ?**

📌 **Comparison Table**

| **Feature**           | **Subplot**                                | **SubFigure**                                   |
|-----------------------|---------------------------------------------|-------------------------------------------------|
| **Concept**           | A single plot inside a figure               | A grouping container inside a figure            |
| **Nesting**           | Cannot contain other subplots               | Can contain its own subplots                    |
| **Layout Flexibility**| Shared across all subplots                  | Each subfigure can have independent layout      |
| **Use Case**          | Simple grid of plots                        | Complex layouts (e.g., mixed grid types)        |
| **Introduced In**     | Available from the beginning                | Matplotlib 3.4+                                 |
| **Accessed via**      | `plt.subplot()`, `plt.subplots()`           | `fig.subfigures()`                              |

**🧠 Analogy:**

- Figure = Canvas
- Subplot = Drawings on the canvas (individual charts)
- SubFigure = Panels within the canvas, each with its own group of charts

**🎯 When to Use What?**

| **Situation**                              | **Use**                               |
|-------------------------------------------|----------------------------------------|
| Simple 2x2 or 3x1 plots                    | ✅ `subplots()`                         |
| Need separate layout/styling per group    | ✅ `subfigures()`                       |
| Want separate titles per section          | ✅ `subfigures()`                       |
| Working with older Matplotlib             | ✅ `subplots()` (for compatibility)     |



````python
# now returning to the actual stuffs
plot_object = plt.subplots()
fig, ax = plot_object
print(plot_object)
ax.plot([1,2,3],[1,2,3])
````

### The subplot2grid command

There is another way to make subplots using a grid-like format:

subplot2grid() lets you place a subplot anywhere on a flexible grid, and control its size (rowspan and colspan). It's much more powerful than the basic subplot() function.

**✅ Syntax:**

plt.subplot2grid(shape, loc, rowspan=1, colspan=1)

**🔍 Parameters:**

| **Parameter** | **Description**                                                                 |
|---------------|---------------------------------------------------------------------------------|
| `shape`       | Tuple `(nrows, ncols)` — defines the total grid layout                         |
| `loc`         | Tuple `(row_index, col_index)` — the starting cell of your subplot             |
| `rowspan`     | *(Optional)* How many rows the subplot should span (default is `1`)            |
| `colspan`     | *(Optional)* How many columns the subplot should span (default is `1`)         |


````python
plt.figure(figsize=(14, 6))

ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
ax4 = plt.subplot2grid((3,3), (2,0))
ax5 = plt.subplot2grid((3,3), (2,1))
````

#### 🧠 What's Going On Here?  
**🔹 plt.figure(figsize=(14, 6))**
- This creates a new figure (canvas) with a width of 14 inches and height of 6 inches.
- This larger canvas gives space for multiple plots.

#### 🔳 Grid Definition:
- You're working with a grid of shape (3, 3):
```
+-------+-------+-------+
| (0,0) | (0,1) | (0,2) |
+-------+-------+-------+
| (1,0) | (1,1) | (1,2) |
+-------+-------+-------+
| (2,0) | (2,1) | (2,2) |
+-------+-------+-------+
```
#### ✅ ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
- Starts at cell (0, 0) (top-left)
- Spans all 3 columns in the first row
- Takes the entire top row horizontally
```
+---------------------------+
|           ax1             |
+-------+-------+-----------+
```

#### ✅ ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
- Starts at (1, 0) (second row, first column)
- Spans 2 columns → (1,0) and (1,1)
- Occupies the left 2/3 of the second row
```
+-------+-------+----------+
|  ax2  |       |          |
+-------+-------+----------+
```

#### ✅ ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
- Starts at (1, 2) (second row, third column)
- Spans 2 rows vertically → (1,2) and (2,2)
- Occupies the rightmost column, from middle to bottom
```

+-------+-------+----------+
|       |       |          |
+-------+-------+   ax3    |
|       |       |          |
+-------+-------+----------+
```
#### ✅ ax4 = plt.subplot2grid((3,3), (2,0))
Bottom-left subplot (third row, first column)

#### ✅ ax5 = plt.subplot2grid((3,3), (2,1))
Bottom-middle subplot (third row, second column)

#### 🔲 Final Layout Overview:
```
+------------------------------+
|             ax1              |  ← (row 0), colspan=3
+----------+----------+--------+
|   ax2    |          | ax3    |  ← ax2: (row 1, col 0-1), ax3: rowspan=2
+----------+----------+        |
|  ax4     |   ax5    |        |  ← (row 2, col 0-1)
+----------+----------+--------+
```

### Scatter Plot

#### 🎯 What is a Scatter Plot?
A **scatter plot** is a type of plot that displays individual data points on a two-dimensional graph.  
Each point represents one observation with two variables — one mapped to the **x-axis** and the other to the **y-axis**.

---

#### 🧠 Purpose:
To **visualize the relationship** (correlation, trend, or pattern) between two numerical variables.  
Commonly used in **exploratory data analysis (EDA)** and **regression tasks**.

---

#### 🔍 Key Parameters of `plt.scatter(x, y, c, s, marker, alpha, label)`

| **Parameter** | **Description**                                                                 |
|---------------|---------------------------------------------------------------------------------|
| `x`, `y`      | The coordinates of the points                                                   |
| `c`           | Color of points (can be one color or array-like for coloring based on data)     |
| `s`           | Size of points                                                                 |
| `marker`      | Shape of each point (e.g., circle `'o'`, square `'s'`, triangle `'^'`, etc.)    |
| `alpha`       | Transparency level (from `0` transparent to `1` opaque)                         |
| `label`       | Label used when displaying a legend                                             |


````python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N) # this represent different colors
area = np.pi * (20 * np.random.rand(N))**2  # 0 to 15 point radii

# explaination 
# =====================================================================
# np.random.rand(N) → 50 random values from 0 to 1.
# 20 * np.random.rand(N) → scales that to values between 0 and 20.
# (20 * rand)**2 → squares the value (because scatter size is in area, not radius).
# np.pi * (...) → multiplies by π to get the area of a circle: A = πr².
````

````python
plt.figure(figsize=(14, 6))

plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Spectral')
plt.colorbar()

plt.show()
````

**🎨 What is cmap?**
- cmap stands for colormap — it maps numerical values (like intensity, temperature, probability, etc.) to colors in a gradient.
- It’s used in visualizations where colors represent data values — like in heatmaps, contour plots, and scatter plots with c (color array).
 
**📚 Popular Colormaps:**
| **Name**     | **Description**                    | **Type**      |
|--------------|------------------------------------|---------------|
| `'viridis'`  | Smooth, perceptually uniform       | Sequential    |
| `'plasma'`   | Bright, high contrast              | Sequential    |
| `'inferno'`  | Dark, dramatic                     | Sequential    |
| `'coolwarm'` | Blue to red                        | Diverging     |
| `'Spectral'` | Rainbow-like                       | Diverging     |
| `'Greys'`    | Black to white                     | Sequential    |
| `'jet'`      | Old-school rainbow *(use sparingly)* | Sequential  |

````python
# 🧱 fig.add_subplot() in Matplotlib
# The add_subplot() method is used to add a subplot (axes object) to a figure manually
# — giving you control over where and how your plots are placed.

fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(1,2,1)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel1')
plt.colorbar()

ax2 = fig.add_subplot(1,2,2)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel2')
plt.colorbar()

plt.show()
````

### KDE (kernel density estimation)

#### 📈 What is KDE (Kernel Density Estimation)?

**Kernel Density Estimation** is a non-parametric way to estimate the **probability density function (PDF)** of a continuous variable.

---

#### 🎯 In simple terms:
- KDE helps you **smooth out a histogram**.
- It gives you a **continuous curve** that shows the **distribution of data**.
- It’s great for **understanding the shape** of your data — like where values are concentrated.

---

#### 🔍 Why Use KDE Over Histograms?

| Feature           | Histogram             | KDE                           |
|-------------------|------------------------|-------------------------------|
| Type              | Discrete (bars)        | Continuous (smooth curve)     |
| Bin-sensitive?    | Yes                    | No (uses bandwidth instead)   |
| Visual smoothness | Blocky                 | Smooth                        |
| Underlying PDF    | Approximated with bins | Approximated with kernel sums |

---

#### 🛠️ How It Works (Conceptually):
- Each data point is replaced with a **kernel** — typically a **Gaussian (bell curve)**.
- These curves are then **summed** to create a **smooth estimate** of the distribution.

---

#### ✍️ KDE in Python (using Seaborn):

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(loc=0, scale=1, size=1000)

sns.kdeplot(data, fill=True, color='skyblue')
plt.title("KDE Plot")
plt.show()
```

#### 🧠 Key KDE Parameters

| Parameter    | Description                                                    |
|--------------|----------------------------------------------------------------|
| `kernel`     | Type of smoothing function (Gaussian is most common)           |
| `bandwidth`  | Controls the smoothness; smaller values = more wiggly curves   |
| `bw_adjust`  | Scales the default bandwidth to make the plot more/less smooth |
| `shade`/`fill` | Whether to fill the area under the KDE curve                 |


````python
from scipy.stats import gaussian_kde
# This line uses Kernel Density Estimation (KDE) to estimate the probability density function (PDF) of the data in values.
density = gaussian_kde(values)
density
````

````python
plt.subplots(figsize=(12, 6))

values2 = np.linspace(min(values)-10, max(values)+10, 100)

plt.plot(values2, density(values2), color='#FF7F00')
plt.fill_between(values2, 0, density(values2), alpha=0.5, color='#FF7F00')
plt.xlim(xmin=-5, xmax=5)

plt.show()
````

#### values2 = np.linspace(min(values)-10, max(values)+10, 100)
- This line creates a smooth, evenly spaced range of 100 values that go a bit beyond the minimum and maximum of your dataset (values), allowing for a wider, cleaner KDE or plot visualization.

✅ **Explanation, Step-by-Step**

| Component              | Description                                                              |
|------------------------|--------------------------------------------------------------------------|
| `min(values)`          | The smallest value in your dataset                                       |
| `max(values)`          | The largest value in your dataset                                        |
| `min(values) - 10`     | Moves the left boundary 10 units left (for extra padding in the plot)    |
| `max(values) + 10`     | Moves the right boundary 10 units right                                  |
| `np.linspace(a, b, 100)` | Generates 100 evenly spaced numbers between `a` and `b`                 |
| `values2 = ...`        | Stores that range in the variable `values2`                              |

✅ **Explanation (Line-by-Line):**

| Component           | Description |
|---------------------|-------------|

| `Y2`                | These are the heights of the bars you’re plotting on top of the `Y` bars (stacked). |
| `width=barWidth`    | Controls the width of the bars — usually the same as the one used for `Y` to align them perfectly. |
| `color='#e17055'`   | Fills the bars with a reddish-orange color (warm tone 🍊). |
| `bottom=Y`          | 🧠 This is the key! It tells Matplotlib to start the `Y2` bars **on top of** the `Y` bars, not from zero. This makes the chart **stacked**. |
| `label='Label Y2'`  | Adds a label for the legend, so this layer of the stack is named `Label Y2`. |

#### 📦 What is `plt.boxplot()`?

`plt.boxplot(values)` is a Matplotlib function used to create a **box-and-whisker plot**, a standardized way of displaying the **distribution** of a dataset based on five key summary statistics and identifying **outliers**.

---

#### 🧠 What a Boxplot Shows (Anatomy of a Boxplot)

Imagine your data spread out. A boxplot compresses all of this into a simple box with lines that tell you the **story of your data’s spread**.

| Component           | Description                                                                 |
|---------------------|-----------------------------------------------------------------------------|
| **Minimum**         | The smallest data point (excluding outliers)                                |
| **Q1 (25th percentile)** | 25% of the data falls below this value                              |
| **Median (Q2)**     | 50% of the data falls below this point — central tendency                   |
| **Q3 (75th percentile)** | 75% of the data falls below this value                              |
| **Maximum**         | Largest data point (excluding outliers)                                     |
| **Outliers**        | Values that fall outside 1.5 × IQR (interquartile range) from Q1 or Q3      |


#### 📌 Outlier Detection Formula (Using IQR)

Outliers are detected using the **Interquartile Range (IQR)** as follows:

#### Step 1: Compute IQR  
**IQR = Q3 − Q1**
#### Step 2: Compute Bounds  
- **Lower Bound**: **Q1 − 1.5 × IQR**
- **Upper Bound**: **Q3 + 1.5 × IQR**

#### Step 3: Identify Outliers  
Any data point **less than the lower bound** or **greater than the upper bound** is considered an outlier.
